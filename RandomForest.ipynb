```python
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score


# Split the data
X_train, X_test, y_train_log, y_test_log = train_test_split(X, y_log, test_size=0.3, random_state=42)

# Initialize Random Forest
rf = RandomForestRegressor(random_state=42, n_estimators=100)

# Fit on train
rf.fit(X_train, y_train_log)

# Predict on test (log scale)
y_pred_log = rf.predict(X_test)

# Convert predictions back to original scale
y_pred = np.exp(y_pred_log)
y_test = np.exp(y_test_log)

# Evaluate on original scale
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE (original scale): {rmse:.4f}")
print(f"Test R-squared (original scale): {r2:.4f}")

# Cross-validation (5 folds) on log scale
cv_neg_mse = cross_val_score(rf, X, y_log, cv=5, scoring='neg_mean_squared_error')
cv_rmse_log = np.sqrt(-cv_neg_mse)
cv_r2 = cross_val_score(rf, X, y, cv=5, scoring='r2')

print(f"Cross-validated RMSE (log scale): {cv_rmse_log.mean():.4f} Â± {cv_rmse_log.std():.4f}")
print(f"Cross-validated R-squared (log scale): {cv_r2.mean():.4f} Â± {cv_r2.std():.4f}")
```
